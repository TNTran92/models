```python
import pandas as pd
import os
import shutil
```

```python
parent_dir = "/mnt/sda1/Backup/result/parquet_jpg"

data_files = [os.path.join(parent_dir, item) for item in os.listdir(parent_dir)]

df = pd.concat((pd.read_parquet(f, engine = 'pyarrow') for f in data_files))
```

```python
final_result = df[['name', 'timestamp', 'inferred', 'time(ms)', 'pred_1', 'pred_1_score']]
#final_result[final_result.pred_1_score > 0.5].sort_values(by=['pred_1_score'], ascending=False).head(50)
final_result = final_result[final_result.pred_1_score > 0.5]
result_dict = final_result.to_dict(orient='records')
```

```python
source_dir = "/mnt/sda1/Backup/jpg"
target_dir = "/mnt/sda1/Backup/jpg_filtered"




final_result = df[['name', 'timestamp', 'inferred', 'time(ms)', 'pred_1', 'pred_1_score']]
#final_result[final_result.pred_1_score > 0.5].sort_values(by=['pred_1_score'], ascending=False).head(50)
final_result = final_result[final_result.pred_1_score > 0.5]
result_dict = final_result.to_dict(orient='records')

def move_img_after_detection(source_dir, target_dir, result_dict_list):
    for item in result_dict_list:
        try:
            filename = item['name']
            pred_1 = item['pred_1']
            # put image to the right dir
            # Make sure dir for the class exist
            dest_path = os.path.join(target_dir, pred_1)
            os.makedirs(dest_path, exist_ok=True)
            source_path = os.path.join(source_dir, filename)
            shutil.move(source_path, dest_path)
            print(f"Moved {filename}")
        except Exception as e:
                print(f"Can't move {item['name']}")


move_img_after_detection(source_dir, target_dir, result_dict)
```
