```python
import pandas as pd
import os
```

```python
parent_dir = "/mnt/sda1/Backup/parquet_jpg"

data_files = [os.path.join(parent_dir, item) for item in os.listdir(parent_dir)]

df = pd.concat((pd.read_parquet(f, engine = 'pyarrow') for f in data_files))
```

```python
final_result = df[['name', 'timestamp', 'inferred', 'time(ms)', 'pred_1', 'pred_1_score']]
final_result[final_result.pred_1_score > 0.5].sort_values(by=['pred_1_score'], ascending=False).head(50)
```
